# Data

The data in several of our notebooks are taken from the jean.dat dataset available from the Stanford GraphBase at https://www-cs-faculty.stanford.edu/~knuth/sgb.html. This data gives character scene incidence information from the novel Les Miserables by Victor Hugo.

```python
scenes = {
    0: ('FN', 'TH'),
    1: ('TH', 'JV'),
    2: ('BM', 'FN', 'JA'),
    3: ('JV', 'JU', 'CH', 'BM'),
    4: ('JU', 'CH', 'BR', 'CN', 'CC', 'JV', 'BM'),
    5: ('TH', 'GP'),
    6: ('GP', 'MP'),
    7: ('MA', 'GP')
}

H = hnx.Hypergraph(scenes)
```

# Visualization

## Two Column Bipartite

```python
draw_bipartite.draw_bipartite_using_euler(H)
```

## Advanced Visualization Parameters

### Color
Pass an array of matplotlib colors or a dictionary mapping edges to a color to assign color to individual edges. You can also pass a single string to assign a specific color for all edges.

```python
edge_color = lambda e: plt.cm.tab10(e)
lighter_edge_color = lambda e: edge_color(e) + np.array([0, 0, 0, -.5])

kwargs = {
    'edges_kwargs': {
        'edgecolors': edge_color,
        'facecolors': lighter_edge_color,
        'color': lighter_edge_color
    }
}

draw_bipartite.draw_bipartite_using_euler(H, **kwargs)
```

### Edge order
Set edge order by passing an array of edges in the order you would like to see the edges drawn

```python
draw_bipartite.draw_bipartite_using_euler(H, **kwargs)
```

### Font

Font attributes can be individually set by passing in a dictionary mapping edges to values using `edge_labels_kwargs`

```python
edge_labels_kwargs={
    'fontsize': lambda v: 36 if v == 1 else 12
}

draw_bipartite.draw_bipartite_using_euler(H, edge_labels_kwargs=edge_labels_kwargs, **kwargs)
```

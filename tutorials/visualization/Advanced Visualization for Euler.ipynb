```python
import matplotlib.pyplot as plt
import numpy as np

import hypernetx as hnx
```

# Data

The data in several of our notebooks are taken from the jean.dat dataset available from the Stanford GraphBase at https://www-cs-faculty.stanford.edu/~knuth/sgb.html. This data gives character scene incidence information from the novel Les Miserables by Victor Hugo.

```python
scenes = {
    0: ('FN', 'TH'),
    1: ('TH', 'JV'),
    2: ('BM', 'FN', 'JA'),
    3: ('JV', 'JU', 'CH', 'BM'),
    4: ('JU', 'CH', 'BR', 'CN', 'CC', 'JV', 'BM'),
    5: ('TH', 'GP'),
    6: ('GP', 'MP'),
    7: ('MA', 'GP')
}

H = hnx.Hypergraph(scenes)
```

# Visualization

## Basic Euler Visualization
Use the default drawing tool to visualize `H`. This renders an Euler diagram of the hypergraph where vertices are black dots and hyper edges are convex shapes containing the vertices belonging to the edge set. It is not always possible to render a "correct" Euler diagram for an arbitrary hypergraph. This technique will lead to false positives, cases where a hyper edge incorrectly contains a vertex not belonging to its set.

```python
hnx.draw(H)
```

## A note on random seeds
Many graph layout algorithms use a random initial position of nodes, causing the output to vary. This can be useful when you want to try multiple times to generate a good layout, but troublesom for reproducability. The default `hypernetx` layout algorithm is `nx.spring_layout` which takes a `seed` parameter determining the inital random positioning of the vertices. Passing a specific value across multiple renderings will produce consistent layouts.

```python
kwargs = dict(layout_kwargs={'seed': 39})

plt.figure(figsize=(10, 5))

plt.subplot(131)
hnx.draw(H, **kwargs)

plt.subplot(132)
plt.title('Identical to left')
hnx.draw(H, **kwargs)

plt.subplot(133)
plt.title('No random seed,\ndifferent from left')
hnx.draw(H)
```

## Size-aware Layouts
It can be useful to have the layout algorithm be aware of the size of node and edges in the hypergraph. For example, if `node_radius` is being used, and some nodes are very large, they may overlap with other nodes if the layout algorithm is not aware of this. Or, if some nodes or edges are highly connected, it may be useful to increase the space around those nodes to help improve the drawing quality. We have added a layout algorithm, `layout_with_radius`, that can take hyper node and edge sizes into account.

```python
from hypernetx.drawing.util import layout_with_radius

# calculate the radius of a hyper node or edge based on its membership size
node_and_edge_radius_func = lambda v: len(H.nodes[v]) if v in H.nodes else len(H.edges[v])

hnx.draw(
    H,
    node_radius=node_and_edge_radius_func, # determines the node size shown in the visualization
    layout=layout_with_radius,
    layout_kwargs=dict(
        node_and_edge_radius=node_and_edge_radius_func # determines the node size in the layout
    )
)
```

The example below illustrates the difference between the original layout (left) and the new radius-aware layout (right). The position calculated by the original layout is used as the initial position for the new layout algorithm so that the drawings are as close as possible to help highlight their differences

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.title('Original Layout\n(using networkx.spring_layout)')
pos = hnx.draw(
    H,
    node_radius=node_and_edge_radius_func,
    return_pos=True # ask for the calculated position back
)

plt.subplot(122)
plt.title('Radius Aware Layout\n(using networkx.kamada_kawai_layout)')
hnx.draw(
    H,
    node_radius=node_and_edge_radius_func,
    layout=layout_with_radius,
    layout_kwargs=dict(
        node_and_edge_radius=node_and_edge_radius_func,
        pos=pos # allows the layout algorithm to use the spring layout (left) as its initial position
    )
)
```

## Advanced Visualization Parameters

```python
from collections import defaultdict 
import networkx as nx

colors = defaultdict(lambda: plt.cm.tab10(len(colors)%10))

def get_node_color(v):
    return colors[v]

def get_cell_color(e):
    return get_node_color(e[1])

hnx.draw(
    H,
    with_additional_edges=nx.DiGraph(H.incidences.items),
    edges_kwargs={'edgecolors': 'black'},
    nodes_kwargs={'color': get_node_color},
    additional_edges_kwargs={'edge_color': get_cell_color},
    edge_labels_on_edge=False, edge_label_alpha=1
)
```

```python
import pandas as pd
### A very busy hypergraph

filepath = '../data/newsgroups-topics.csv'

## uncomment the filepath below to run in Colab
# filepath = 'https://raw.githubusercontent.com/pnnl/HyperNetX/master/tutorials/data/newsgroups-topics.csv'

df = pd.read_csv(filepath, index_col=0)
df.index = df.index.map(lambda s: '.'.join(s.split('.')[-2:]))

H = hnx.Hypergraph.from_incidence_dataframe(df)
# create functions for mapping hyper edges/weights to colors
norm = plt.Normalize(0, H.incidence_matrix().max().max())
cmap = plt.cm.Greens

def get_cell_color(e):
    return cmap(norm(H.incidences[e].weight))

plt.figure(figsize=(12, 12))
hnx.draw(
    H,
    layout=nx.kamada_kawai_layout,
    with_additional_edges=H.bipartite(),
    edges_kwargs={
        'edgecolors': 'darkgray',
        'facecolors': (.65, .65, .65, .15)
    },
    additional_edges_kwargs={
        'edge_color': get_cell_color,
        'width': 4,
    },
    edge_labels_on_edge=False, edge_label_alpha=1,
    contain_hyper_edges=True
)
```

```python
## Let's take a subhypergraph focusing on the most interesting edges

threshold = 0.1

# filter out singletons
newdf = df[df.columns[(df >= threshold).sum(axis=0) > 1]]

# filter out small weights
newdf[newdf < threshold] = None

# construct hypergraph
H2 = hnx.Hypergraph.from_incidence_dataframe(newdf.T)
norm = plt.Normalize(0, H2.incidence_matrix().max().max())
cmap = plt.cm.Greens

def get_cell_color(e):
    return cmap(norm(H2.incidences[e].weight))

plt.figure(figsize=(12, 12))
hnx.draw(
    H2,
    layout=nx.kamada_kawai_layout,
    with_additional_edges=H2.bipartite(),
    edges_kwargs={
        'edgecolors': 'darkgray',
        'facecolors': (.65, .65, .65, .15)
    },
    additional_edges_kwargs={
        'edge_color': get_cell_color,
        'width': 4,
    },
    edge_labels_on_edge=False, edge_label_alpha=1,
    contain_hyper_edges=True
)
```
